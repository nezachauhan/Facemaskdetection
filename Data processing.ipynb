```python
# Dataset Varification

import os
import numpy as np
import shutil
import random

root_dir = 'data'
# Making Directories
posCls = '/with_mask'
negCls = '/without_mask'
os.makedirs(root_dir +'/train' + posCls)
os.makedirs(root_dir +'/test'+ posCls)
os.makedirs(root_dir +'/train' + negCls)
os.makedirs(root_dir +'/test'+ negCls)

# Creating partitions of the data after shuffling
currentCls = posCls
# Folder to copy images from
src = "data"+currentCls 

allFileNames = os.listdir(src)
np.random.shuffle(allFileNames)
# Split data into train,test folders
train_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                                          [int(len(allFileNames)*0.625), int(len(allFileNames)*0.82)])


train_FileNames = [src+'/'+ name for name in train_FileNames.tolist()]
test_FileNames = [src+'/' + name for name in test_FileNames.tolist()]

print('Total images: ', len(allFileNames))
print('Training: ', len(train_FileNames))
print('Testing: ', len(test_FileNames))

# Copy-paste images
for name in train_FileNames:
    shutil.copy(name, "Data/train"+currentCls)

for name in test_FileNames:
    shutil.copy(name, "Data/test"+currentCls)

# For without mask data
currentCls1 = negCls
src1 = "Data"+currentCls1

allFileNames1 = os.listdir(src1)
np.random.shuffle(allFileNames1)
train_FileNames1, test_FileNames1 = np.split(np.array(allFileNames1),
                                                          [int(len(allFileNames1)*0.625), int(len(allFileNames1)*0.82)])


train_FileNames1 = [src1+'/'+ name for name in train_FileNames1.tolist()]
test_FileNames1 = [src1+'/' + name for name in test_FileNames1.tolist()]

print('Total images: ', len(allFileNames1))
print('Training: ', len(train_FileNames1))
print('Testing: ', len(test_FileNames1))

for name in train_FileNames1:
    shutil.copy(name, "Data/train"+currentCls1)

for name in test_FileNames1:
    shutil.copy(name, "Data/test"+currentCls1)
```

```python
import os
from PIL import Image 
import PIL
import os.path
# path for mask folders
f = 'data/train/with_mask' 
for file in os.listdir(f):
    f_img = f+"/"+file
    img = Image.open(f_img)
    img = img.resize((150,150)) # resize image into 150 x 150
    img.save(f_img)
# path for without mask images
nf = 'data/train/without_mask'
for file in os.listdir(nf):
    f_img = nf+"/"+file
    img = Image.open(f_img)
    img = img.resize((150,150))
    img.save(f_img)

folder_images = "data/train"
size_images = dict()

# to change size for all images from different folders
for dirpath, _, filenames in os.walk(folder_images):
    for path_image in filenames:
        image = os.path.abspath(os.path.join(dirpath, path_image))
        with Image.open(image) as img:
            width, heigth = img.size
            size_images[path_image] = {'width': width, 'heigth': heigth}

print(size_images)
```
